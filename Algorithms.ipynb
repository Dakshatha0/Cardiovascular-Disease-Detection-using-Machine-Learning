```python
import pandas as pd
import os

# Define the base directories where CSV files are located for each category
base_dirs = {
    'normal': '/content/drive/MyDrive/gwbz3fsgp8-3/Normal Person ECG Images (284x12=3408)',
    'abnormal': '/content/drive/MyDrive/gwbz3fsgp8-3/ECG Images of Patient that have abnormal heartbeat (233x12=2796)',
    'MI': '/content/drive/MyDrive/gwbz3fsgp8-3/ECG Images of Myocardial Infarction Patients (240x12=2880)',
    'PMI': '/content/drive/MyDrive/gwbz3fsgp8-3/ECG Images of Patient that have History of MI (172x12=2064)'
}

# List to store DataFrames for each lead from all folders
lead_dfs = {lead_no: [] for lead_no in range(1, 14)}  # Initialize empty lists for each lead (1-13)

# Iterate over each category and its corresponding base directory
for category, base_dir in base_dirs.items():
    # Iterate over all files in the current base directory
    for file_name in os.listdir(base_dir):
        if file_name.startswith('scaled_data_1D_') and file_name.endswith('.csv'):
            # Extract lead number from the file name
            lead_no = int(file_name.split('_')[3].split('.')[0])

            # Construct the full path to the CSV file
            file_path = os.path.join(base_dir, file_name)

            # Read CSV file into a DataFrame and append to the corresponding lead list
            try:
                df = pd.read_csv(file_path)
                lead_dfs[lead_no].append(df)
            except FileNotFoundError:
                print(f"CSV file not found: {file_path}")

# Iterate over lead numbers (1-13) and combine DataFrames for each lead
for lead_no, dfs in lead_dfs.items():
    if dfs:
        # Concatenate DataFrames along rows for the current lead
        combined_df = pd.concat(dfs, ignore_index=True)

        # Define path to save the combined CSV file for the current lead
        combined_csv_path = f'/content/drive/MyDrive/gwbz3fsgp8-3/Combined_CSV_Lead_{lead_no}.csv'

        # Save the combined DataFrame to a CSV file
        combined_df.to_csv(combined_csv_path, index=False)

        print(f"Combined CSV file for Lead {lead_no} saved to:", combined_csv_path)
    else:
        print(f"No CSV files found for Lead {lead_no}")
```

```python
import pandas as pd

# Specify the path to the CSV file containing the combined data for lead 1
file_path = '/content/drive/MyDrive/gwbz3fsgp8-3/Combined_CSV_Lead_1.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the unique values in the 'Target' column
unique_targets = df['Target'].unique()
print("Unique targets in 'Target' column:", unique_targets)
```

```python
# Check the columns in the DataFrame
print(df.columns)

# Drop the index column if it's present
if 'Unnamed: 0' in df.columns:
    df.drop(columns=['Unnamed: 0'], inplace=True)
else:
    print("Column 'Unnamed: 0' not found in DataFrame.")
```

```python
#convert Target column values as Numeric using ngroups
encode_target_label = df.groupby('Target').ngroup().rename("target").to_frame()
test_final  = df.merge(encode_target_label, left_index=True, right_index=True)
test_final.drop(columns=['Target'],inplace=True)
test_final
```

**DIMENSIONALITY REDUCTION**

```python
import pandas as pd
from sklearn.decomposition import PCA

# Assuming test_final is your DataFrame with encoded labels
# Perform PCA and choose components as 100
pca = PCA(n_components=100)
x_pca = pca.fit_transform(test_final.iloc[:, 0:-1])  # Perform PCA on all columns except the target column
x_pca = pd.DataFrame(x_pca)

# Calculate the variance explained by principal components
explained_variance = pca.explained_variance_ratio_
print('Variance of each component:', explained_variance)
print('\nTotal Variance Explained:', round(sum(explained_variance) * 100, 2))  # Total variance explained in percentage

# Construct the result DataFrame with PCA components and target variable
pca_df = pd.DataFrame(data=x_pca)
target = pd.Series(test_final['target'], name='target')  # Extract target variable from test_final
result_df = pd.concat([pca_df, target], axis=1)  # Concatenate PCA components and target variable

# Display the result DataFrame
print(result_df)
```

```python
# Display the unique values in the 'target' column
unique_targets = result_df['target'].unique()
print("Unique targets in 'target' column:", unique_targets)
```

**ML MODELS FOR ONE LEAD**

K-Nearest Neighbours

```python
# Import the necessary modules for ML model
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report

# Setup the pipeline steps:
steps = [('knn', KNeighborsClassifier())]

# Create the pipeline: pipeline
pipeline = Pipeline(steps)

# have paased less range value of hyperparamter since i'm using free tier version of google colab.
k_range = list(range(1, 9))
parameters = dict(knn__n_neighbors=k_range)

#input
X = result_df.iloc[:,0:-1]

#target
y=result_df.iloc[:,-1]

# Create train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4,random_state=42)

#increasing cv score takes lot of time in gooogle colab, so kept it just 2.
cv = GridSearchCV(pipeline,parameters,cv=2)

cv.fit(X_train,y_train)

# Predict the labels of the test set: y_pred
y_pred = cv.predict(X_test)

Knn_Accuracy = cv.score(X_test, y_test)

# Compute and print metrics
print("Accuracy: {}".format(Knn_Accuracy*100))
print(classification_report(y_test, y_pred))
print("Tuned Model Parameters: {}".format(cv.best_params_))
```

ENSEMBLE LEARNING

```python
from sklearn.ensemble import VotingClassifier

# Define the base models
model1 = ('knn', KNeighborsClassifier())
model2 = ('lr', LogisticRegression(max_iter=5000))  # Increase max_iter
model3 = ('svm', SVC())

# Create the ensemble classifier
ensemble = VotingClassifier(estimators=[model1, model2, model3])

# Fit the ensemble model
ensemble.fit(X_train, y_train)

# Predict the labels of the test set
ensemble_pred = ensemble.predict(X_test)

# Compute accuracy
ensemble_accuracy = ensemble.score(X_test, y_test)
print("Ensemble Accuracy:", ensemble_accuracy*100)

# Print classification report
print(classification_report(y_test, ensemble_pred))

from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

# Calculate accuracy
ensemble_accuracy = accuracy_score(y_test, ensemble_pred)
print("Ensemble Accuracy:", ensemble_accuracy*100, "%")

# Calculate precision, recall, and F1-score
precision = precision_score(y_test, ensemble_pred, average='weighted')
recall = recall_score(y_test, ensemble_pred, average='weighted')
f1 = f1_score(y_test, ensemble_pred, average='weighted')

# Print precision, recall, and F1-score
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
